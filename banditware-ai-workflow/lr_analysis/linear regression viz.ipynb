```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import plotly
import seaborn as sns
```

```python
# import data
df1 = pd.read_csv('lr_1.csv')
df2 = pd.read_csv('lr_2.csv')
```

```python
# retrieve duration columns
duration_col1 = df1['Duration'].str.replace('s', '').astype(float)
duration_col2 = df2['Duration'].str.replace('s', '').astype(float)
duration_df = pd.DataFrame({'duration_all': duration_col1, 'duration_area_only': duration_col2})
```

```python
# retrieve RMSE scores
rmse_col1 = df1['training_root_mean_squared_error']
rmse_col2 = df2['training_root_mean_squared_error']
rmse_df = pd.DataFrame({'rmse': rmse_col1, 'rmse_area_only': rmse_col2})
```

```python
duration_df.plot(kind='box', title='duration (seconds) for linear regression recommenders', grid=True) 
```

```python
rmse_df.plot(kind='box', title='RMSE scores for linear regression recommenders', grid=True) 
```
