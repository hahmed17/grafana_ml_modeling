```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import plotly
import seaborn as sns
```

```python
# bp3d
```

```python
# import data
df1 = pd.read_csv('matmul_full_data_runs.csv')
df2 = pd.read_csv('matmul_truncated_data_runs.csv')
```

```python
# retrieve duration columns
duration_col1 = df1['Duration'].str.replace('s', '').astype(float)
duration_col2 = df2['Duration'].str.replace('s', '').astype(float)
duration_df = pd.DataFrame({'duration_all': duration_col1, 'duration_area_only': duration_col2})
```

```python
# retrieve RMSE scores
rmse_col1 = df1['training_root_mean_squared_error']
rmse_col2 = df2['training_root_mean_squared_error']
rmse_df = pd.DataFrame({'rmse_all': rmse_col1, 'rmse_truncated': rmse_col2})
```

```python
# retrieve R^2 scores
r2_col1 = df1['training_r2_score']
r2_col2 = df2['training_r2_score']
r2_df = pd.DataFrame({'r2_all': r2_col1, 'r2_truncated': r2_col2})
```

```python
duration_fig = duration_df.plot(kind='box', title='duration (seconds) for linear regression recommenders (matmul)', grid=True).get_figure() 
duration_fig.savefig('../figures/matmul_linreg_duration.png')
```

```python
rmse_fig = rmse_df.plot(kind='box', title='RMSE scores for linear regression recommenders (matmul)', grid=True).get_figure()
rmse_fig.savefig('../figures/matmul_linreg_rmse.png')
```

```python
r2_fig = r2_df.plot(kind='box', title='R-squared scores for linear regression recommenders (matmul)', grid=True).get_figure()
r2_fig.savefig('../figures/matmul_linreg_r2.png')
```
