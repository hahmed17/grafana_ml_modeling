```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import plotly
import seaborn as sns
from matplotlib.ticker import MultipleLocator
```

```python
# bp3d
```

```python
# import data
df1 = pd.read_csv('matmul_full_data_runs.csv')
df2 = pd.read_csv('matmul_truncated_data_runs.csv')
```

```python
# retrieve duration columns
duration_col1 = df1['Duration'].str.replace('s', '').astype(float)
duration_col2 = df2['Duration'].str.replace('s', '').astype(float)
duration_df = pd.DataFrame({'duration_all': duration_col1, 'duration_area_only': duration_col2})
```

```python
# retrieve RMSE scores
rmse_col1 = df1['training_root_mean_squared_error']
rmse_col2 = df2['training_root_mean_squared_error']
rmse_df = pd.DataFrame({'rmse_all': rmse_col1, 'rmse_truncated': rmse_col2})
```

```python
# retrieve R^2 scores
r2_col1 = df1['training_r2_score']
r2_col2 = df2['training_r2_score']
r2_df = pd.DataFrame({'r2_all': r2_col1, 'r2_truncated': r2_col2})
```

```python
plt.rcParams["figure.figsize"] = (6, 4)
```

```python
# Create the plot
ax = duration_df.plot(kind='box', title='Duration (seconds) for linear regression recommenders (matmul)', grid=True)

# Set y-axis ticks at regular intervals (optional, modify as needed)
y_min, y_max = ax.get_ylim()
ax.yaxis.set_major_locator(MultipleLocator(0.5))  # Adjust interval as appropriate for your data

# Adjust layout and save the figure
duration_fig = ax.get_figure()
duration_fig.tight_layout()  # Ensures layout adjustments for clarity
duration_fig.savefig('../figures/matmul_linreg_duration.png')
```

```python
# Create the plot
ax = rmse_df.plot(kind='box', title='RMSE scores for linear regression recommenders (matmul)', grid=True)

# Set y-axis ticks at intervals of 1.0 (optional, adjust based on your data range)
y_min, y_max = ax.get_ylim()
ax.yaxis.set_major_locator(MultipleLocator(1.0))

# Adjust layout and save the figure
rmse_fig = ax.get_figure()
rmse_fig.tight_layout()  # Ensures layout adjustments for clarity
rmse_fig.savefig('../figures/matmul_linreg_rmse.png')
```

```python
# Create the plot
ax = r2_df.plot(kind='box', title='R-squared scores for linear regression recommenders (matmul)', grid=True)

# Set y-axis ticks at intervals of 0.1 (optional, adjust based on your data range)
y_min, y_max = ax.get_ylim()
ax.yaxis.set_major_locator(MultipleLocator(.1))

# Force the ticks to appear
ax.tick_params(axis='y', which='both', bottom=True, top=True, left=True, right=True) 

# Adjust layout and save the figure
r2_fig = ax.get_figure()
r2_fig.tight_layout()  # Ensures layout adjustments for clarity
r2_fig.savefig('../figures/matmul_linreg_r2.png')
```
